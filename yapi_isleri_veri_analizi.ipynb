## 1.2 Fatura Tarihi	değişkeni veri analizi

```python
# Hangi gün kaç farklı satış yapılmış onu gösteriyor.
df["Fatura Tarihi"].value_counts()
```

```python
# Fatura Tarihi değişkenindeki Boş değerleri gösterir.
df["Fatura Tarihi"].isnull().sum()
```

```python
# bunu lambda fonksiyonu ile daha fonksiyonel yazabilirsin.
df["Numara"] = df["Numara"].fillna("bos")
i = 0
while i<len(df["Numara"])-1:
    if df["Numara"][i+1] == "bos":
        df["Numara"][i+1] = df["Numara"][i] 
    
    i +=1
    

df["Fatura Tarihi"] = df["Fatura Tarihi"].fillna("bos")
i = 0
while i<len(df["Fatura Tarihi"])-1:
    if df["Fatura Tarihi"][i+1] == "bos":
        df["Fatura Tarihi"][i+1] = df["Fatura Tarihi"][i] 
    
    i +=1

    
df["Durumu"] = df["Durumu"].fillna("bos")
i = 0
while i<len(df["Durumu"])-1:
    if df["Durumu"][i+1] == "bos":
        df["Durumu"][i+1] = df["Durumu"][i] 
    
    i +=1
    
df["Tür"] = df["Tür"].fillna("bos")
i = 0
while i<len(df["Tür"])-1:
    if df["Tür"][i+1] == "bos":
        df["Tür"][i+1] = df["Tür"][i] 
    
    i +=1
    
    
df["İş Ortağı/İsim"] = df["İş Ortağı/İsim"].fillna("bos")
i = 0
while i<len(df["İş Ortağı/İsim"])-1:
    if df["İş Ortağı/İsim"][i+1] == "bos":
        df["İş Ortağı/İsim"][i+1] = df["İş Ortağı/İsim"][i] 
    
    i +=1
    
df["Tür"] = df["Tür"].fillna("bos")
i = 0
while i<len(df["Tür"])-1:
    if df["Tür"][i+1] == "bos":
        df["Tür"][i+1] = df["Tür"][i] 
    
    i +=1
```
